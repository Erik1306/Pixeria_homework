```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('rgb_image.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
plt.hist(img.ravel(),256,[0,256])
```

```python
img_ravel = np.ravel(img)
unique, counts = np.unique(img_ravel, return_counts=True)
dict_img_ravel = dict(zip(unique, counts))
```

```python
dict_img_ravel[254]
```

```python
s = 0
new_dict = {}
for i in range(254):
 
    s = s+dict_img_ravel[i]
    new_dict[i] = np.round(255*s/sum(dict_img_ravel.values()))
print(new_dict)
        
    
```

```python
plt.bar(list(new_dict.keys()), new_dict.values(), color='b')
plt.show()
```
