```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
grayscale_img = cv.imread('grayscale_to_binary.png')
```

```python
threshold = np.mean(grayscale_img)
```

```python
threshold
```

```python
tres,gray = cv.threshold(grayscale_img,threshold, 256, cv.THRESH_BINARY )
```

```python
plt.imshow(gray)
```

```python
grayscale_img.shape
```

```python
for i in range(3):
    for j in range(1848):
        for k in range(1048):
            if grayscale_img[:,:,i][:,j][k] > threshold:
                grayscale_img[:,:,i][:,j][k] = 255
            else:
                grayscale_img[:,:,i][:,j][k] = 0
                
```

```python
binary_img = cv.imwrite('binary.jpg',grayscale_img)
```
