```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
rgb_image = cv.imread('rgb_image.jpg')
```

```python
plt.imshow(rgb_image, cmap = 'gray')
plt.title('Original')
plt.show()
```

```python
plt.imshow(red, cmap = 'gray')
plt.title('Red')
plt.show()
```

```python
plt.imshow(green, cmap = 'gray')
plt.title('Green')
plt.show()
```

```python
plt.imshow(blue, cmap = 'gray')
plt.title('Blue')
plt.show()
```
